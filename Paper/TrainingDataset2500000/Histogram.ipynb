```python
import numpy as np
import pandas as pd
import time
import logging

import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 20})
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
```

```python
#Define directory of the input and output files 
DATA_DIR = ''

#Define location to the training and validation dataset
TRAINING_DATASET_FILE_PATH = DATA_DIR+r'1000000LambdaEtaAlignedSpinTrainingDataset.csv'
TrainingBank1 = pd.read_csv(TRAINING_DATASET_FILE_PATH)

#Define location to the training and validation dataset
TRAINING_DATASET_FILE_PATH = DATA_DIR+r'1500000DiffusedLambdaEtaAlignedSpinTrainingDataset.csv'
TrainingBank2 = pd.read_csv(TRAINING_DATASET_FILE_PATH)
```

```python
plt.figure(figsize=(9, 9))
plt.hist(TrainingBank1['match'], bins=50,  alpha=0.5, color='#8855fc', label='Normal Dataset')
plt.hist(TrainingBank2['match'], bins=50,  alpha=0.5, color='#ff3399', label='Diffused Dataset')
plt.xlabel('$\mathcal{M}$')
plt.ylabel('Count')
plt.legend(loc='upper left')
plt.savefig('match_error_histogram_equal.png', dpi=300, bbox_inches='tight')
```
